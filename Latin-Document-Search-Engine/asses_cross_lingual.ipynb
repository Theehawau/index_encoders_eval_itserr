```python
import pandas as pd
from pathlib import Path
import os
from glob import glob
```

```python
cosine_results = list(glob("data_cross/Latin/*/Latin_benchmark_best_cosine_similarities.txt"))

for i in cosine_results:
    print("--------------------------------------------------")
    print(Path(i).parent.name)
    print("--------------------------------------------------")

    df = pd.read_csv(i, sep="\t")
    gr_targets = [f'Greek_Target #{i}' for i in range(0,5)] 
    la_targets = [f'Latin_Target #{i}' for i in range(1,6)]
    
    all_targets = gr_targets + la_targets

    print("Latin Target Metrics:")
    print(df[la_targets].mean(axis=1).describe())

    print("Greek Target Metrics:")
    print(df[gr_targets].mean(axis=1).describe())


    print("All Target Metrics:")
    print(df[all_targets].mean(axis=1).describe())
```
